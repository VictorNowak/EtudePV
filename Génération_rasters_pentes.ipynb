```python
import os
import subprocess
from glob import glob
from tqdm import tqdm

# === Param√®tres ===
input_dir = "DSM"  # dossier avec tes X DSM
output_dir = os.path.join(input_dir, "slope_tiles")
os.makedirs(output_dir, exist_ok=True)

# Liste des fichiers DSM
dsm_files = sorted(glob(os.path.join(input_dir, "*.tif")))

print(f"üì¶ {len(dsm_files)} DSM trouv√©s.")

# Boucle de traitement
for dsm_path in tqdm(dsm_files, desc="Calcul des pentes"):
    # Nom de sortie
    base_name = os.path.basename(dsm_path).replace(".tif", "_slope.tif")
    slope_path = os.path.join(output_dir, base_name)

    # Commande GDAL slope
    cmd = [
        "gdaldem", "slope",
        dsm_path, slope_path,
        "-s", "1.0",        # facteur d'√©chelle (1 si coordonn√©es en m)
        "-compute_edges",
        "-of", "GTiff"
    ]

    # Lancer la commande
    subprocess.run(cmd, check=True)

print(f"‚úÖ Termin√©. Fichiers de pente dans : {output_dir}")
```
