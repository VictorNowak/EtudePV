```python
import os
import subprocess
from glob import glob
from tqdm import tqdm

# === ParamÃ¨tres ===
input_dir = "DSM"  # dossier avec tes X DSM
output_dir = os.path.join(input_dir, "aspect_tiles")
os.makedirs(output_dir, exist_ok=True)

# Liste des fichiers DSM
dsm_files = sorted(glob(os.path.join(input_dir, "*.tif")))

print(f"ðŸ“¦ {len(dsm_files)} DSM trouvÃ©s.")

# Boucle de traitement
for dsm_path in tqdm(dsm_files, desc="Calcul des orientations (aspect)"):
    # Nom de sortie
    base_name = os.path.basename(dsm_path).replace(".tif", "_aspect.tif")
    aspect_path = os.path.join(output_dir, base_name)

    # Commande GDAL aspect
    cmd = [
        "gdaldem", "aspect",
        dsm_path, aspect_path,
        "-zero_for_flat",  # Les zones plates auront une valeur 0
        "-of", "GTiff"
    ]

    # Lancer la commande
    subprocess.run(cmd, check=True)

print(f"âœ… TerminÃ©. Fichiers d'aspect dans : {output_dir}")
```
